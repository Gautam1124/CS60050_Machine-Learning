```python
from ucimlrepo import fetch_ucirepo
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import matplotlib.pyplot as plt
import random
from collections import Counter
import seaborn as sns
```

## Part A

```python
def simulate_expr(k,times,roll_per_time):
    values=[]
    probs=[]
    ans=[]
    for i in range(1,k+1):
        values.append(i)
        if(i==1):
            probs.append(1/(2**(k-1)))
        else:
            probs.append(1/(2**(i-1)))
    
    for i in range(times):
        sum=0
        for j in range(roll_per_time):
            rand_val=random.choices(values,weights=probs)
            sum+=rand_val[0]
        ans.append(sum)
    return values,probs,ans
```

```python
def evaluate(values,probabilities,res,k,times,roll_per_time):
    plt.hist(res)
    plt.show()
    
    num_expr= len(res)
    values_c = Counter(res)
    diff_outcomes = list(values_c.keys())
    probs=[]
    for c in values_c.values():
        probs.append(c/num_expr)
        
    expected_result_sim = sum(res)
    expected_result_theoritical=0
    sum_loc=0
    for (v,p) in zip(values,probabilities):
        sum_loc+=p*v
    expected_result_theoritical=sum_loc*times*roll_per_time
        
    print("<----------------------------------Results--------------------------------------------------->")
    print("Outcomes:", diff_outcomes)
    print("Probabilities:", probs)
    print("SUM using simulation:", expected_result_sim)
    print("Expected value of Sum of events Theoritical:", expected_result_theoritical)

   
```

```python
values1,probs1,res1=simulate_expr(4,1000,4)
evaluate(values1,probs1,res1,4,1000,4)
```

```python
values2,probs2,res2=simulate_expr(4,1000,8)
evaluate(values2,probs2,res2,4,1000,8)
```

```python
values3,probs3,res3=simulate_expr(16,1000,4)
evaluate(values3,probs3,res3,16,1000,4)
```

```python
df = pd.DataFrame(data=X, columns=spambase.metadata.features)
df['spam'] = y

# Choose 5 columns for probability distribution plots
selected_columns = ['word_freq_hp', 'word_freq_make', 'word_freq_internet', 'word_freq_mail', 'word_freq_all']

plt.figure(figsize=(15, 8))

# Plot the probability distribution for each selected column
for i, column in enumerate(selected_columns):
    plt.subplot(2, 3, i + 1)
    sns.histplot(data=df, x=column, stat="probability", discrete=True)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Probability Density')

plt.tight_layout()
plt.show()
```

```python
class Naive_bayes :
    
    def __init__(self) :
        self.no_of_features = 0
        self.mean_list = [[],[]]
        self.var_list = [[],[]]
        self.p = [[],[]]
        self.y_pred = []
        self.no_of_parameters = 0
        
    def fit(self,X_train,y_train) :
        self.no_of_features = X_train.shape[1]
        sum1 = np.sum(y_train,axis=0)
        self.p[1] = sum1/y_train.shape[0]
        self.p[0] = 1-self.p[1]
        for i in range(self.no_of_features) :
            self.mean_list[1].append(np.mean(X_train.loc[(y_train['Class']==1),X_train.columns[i]]))
            self.mean_list[0].append(np.mean(X_train.loc[(y_train['Class']==0),X_train.columns[i]]))
            self.var_list[1].append(np.var(X_train.loc[(y_train['Class']==1),X_train.columns[i]]))
            self.var_list[0].append(np.var(X_train.loc[(y_train['Class']==0),X_train.columns[i]]))
            
        
        self.mean_list[1] = np.array(self.mean_list[1])
        self.mean_list[0] = np.array(self.mean_list[0])  
        self.var_list[1] = np.array(self.var_list[1])
        self.var_list[0] = np.array(self.var_list[0])
        
        self.no_of_parameters = 2*self.no_of_features + 2
        
    def predict(self,X_test) :
        for i in range(X_test.shape[0]) :
            p1 = np.prod((1/np.sqrt(2*np.pi*self.var_list[1]))*np.exp(-((X_test.iloc[i,:]-self.mean_list[1])**2)/(2*self.var_list[1])))
            p0 = np.prod((1/np.sqrt(2*np.pi*self.var_list[0]))*np.exp(-((X_test.iloc[i,:]-self.mean_list[0])**2)/(2*self.var_list[0])))
            
            if (p1*self.p[1] > p0*self.p[0]).all():
                self.y_pred.append(1)
            else :
                self.y_pred.append(0)
        # return np array
        return self.y_pred
```

```python
# Initialize and fit the Naive Bayes classifier
nb_classifier = Naive_bayes()
nb_classifier.fit(X_train, y_train)

# Make predictions on the test data
y_pred = nb_classifier.predict(X_test)
y_pred
```

```python
def print_res(y_test,y_pred):
    y_test= np.ravel(y_test)
    y_pred=np.ravel(y_pred)
    f1 = f1_score(y_test ,y_pred)
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test ,y_pred)
    recal = recall_score(y_test ,y_pred)
    print("accuracy: " , acc , " precision: ",prec," recall: ",recal," F1-score: ",f1)

print_res(y_test,y_pred)
```

```python
X_train_new = np.log(X_train+1)
X_test_new = np.log(X_test+1)
X_train_new
```

```python
nb_classifier_new = Naive_bayes()
nb_classifier_new.fit(X_train_new, y_train)

# Make predictions on the test data
y_pred = nb_classifier_new.predict(X_test_new)
```

```python
print_res(y_test,y_pred)
```
